```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.tradeindia.com/manufacturers/bio-fertilizers.html'
```

```python
HEADERS = ({'User-Agent': 'add_your_user_agent', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage.content
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python
soup
```

```python
links = soup.find_all('a', attrs={'class':'product-title-position'})
```

```python
links
```

```python
links[0]
```

```python
link = links[0].get('href')
```

```python
link
```

```python
new_webpage = requests.get(link, headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
```

```python
new_soup
```

```python
# product name
new_soup.find('h1', attrs={'class':'sc-99c38e0-3 crggkU m-0 Headline3'}).text.strip()
```

```python
# price
new_soup.find('h2', attrs={'class':'sc-99c38e0-3 kpwqmK prPrice Headline3'}).text.strip()
```

```python
new_soup.find('span', attrs={'class':'mr-1'}).text.strip() #rating
```

```python
# company
new_soup.find('h2', attrs={'class':'sc-99c38e0-1 dtjLEg co-name'}).text.strip()
```

```python
# contact name
new_soup.find('p', attrs={'class':'sc-99c38e0-13 bFlIdM m-0 mb-3 Body4R'}).text.strip()
```

```python
# address

elements = new_soup.find_all('p', attrs={'class': 'sc-99c38e0-13 bFlIdM m-0 mb-3 Body4R'})

if len(elements) >= 3:
    third_element_text = elements[2].text.strip()
    print(third_element_text)
else:
    print("There are fewer than three elements with the specified class.")
```
